## Official Google API

### Show response

```python
import requests
import json
import pandas as pd
search=input('Give me a term you want to search, I will show you the result restricted in HK and in 2022\n')
dateRestrict="y[2022]"
mode=input('Do you need exact search or simple search? Input simple or exact in lowercase\n')
pagenum=input('How many pages do you like me to scrap?')
para={'key':'#type your key','cx':'type your id',
        'q':search,'dateRestrict':dateRestrict,
        'gl':'HK', 'exactTerms':search
    }

def to_ex(dataresult,n):
    #convert dataFrame to excel
    if n == 'yes':
        dataresult.to_excel('{} result.xlsx'.format(search))
        print("Your file is stored as {}.xlsx".format(search))
    elif n == 'no':
        return dataresult
        
def get(mode,startnum):
    #get API response
    if mode=='exact':
        return requests.get("https://www.googleapis.com/customsearch/v1?key={}&cx={}&exactTerms={}&dateRestrict={}&gl={}&start={}".format(para['key'],para['cx'],para['exactTerms'],para['dateRestrict'],para['gl'],startnum))
    elif mode=='simple':
        return requests.get("https://www.googleapis.com/customsearch/v1?key={}&cx={}&q={}&dateRestrict={}&gl={}&start={}".format(para['key'],para['cx'],para['q'],para['dateRestrict'],para['gl'],startnum))

def responseToDf(startnum):
    response=get(mode,startnum)
    response_dict=json.loads(response.text)
    outcome={}
    for n in range(len(response_dict)):
        title=response_dict['items'][n]['title']
        link=response_dict['items'][n]['link']
        outcome[title]=link
    df = pd.DataFrame(list(outcome.items()),columns = ['Title','Link'])
    return df

def dfconcat(pagenumber):
    df=responseToDf(1)
    if pagenumber==1:
        return df
    else:
        startindex=[x for x in range(11,(pagenumber)*10+2,10)]
        for startnum in startindex:
            df1=responseToDf(startnum)
            df=pd.concat([df,df1],ignore_index=True)
        return df
        
dfconcat(pagenum)
outputornot=input('Do you need a copy in xlsx file?\n')
to_ex(df,outputornot)
#create function to convert response and join 10 times!!! 19/10/2022
#response=get(mode,1)
#response_dict = json.loads(response.text)

#outcome={}
#for n in range(len(response_dict)):
#    title=response_dict['items'][n]['title']
#    link=response_dict['items'][n]['link']
#    outcome[title]=link
#df = pd.DataFrame(list(outcome.items()),columns = ['Title','Link'])

#google_trend = input('Do you want to see the interest level of {} in HK?'.format(search))
```
